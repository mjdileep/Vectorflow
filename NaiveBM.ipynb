```python
from vectorflow.classical.naive_bm import NaiveBM
import numpy as np
%load_ext autoreload
%autoreload 2
#https://en.wikipedia.org/wiki/Boltzmann_machine
```

```python
n = 100 # Total number of units
v = 20 # Visible vector size 
d=20 # number of training data points
data = np.random.rand(d, v) <0.5 # Training data 
temperature = 4 # For Simulated annealing effect https://en.wikipedia.org/wiki/Simulated_annealing

bm = NaiveBM(n, v, temperature) # Create BM with total n units and v visible units
bm.fit(data) # Fit the BM

# Testing BM by sending fraction of data 
k = 1.5
for each in range(d):
    s = bm.retrieve(data[each,:int(v/k)])
    print("\nInput:",data[each,:int(v/k)])
    print("expected:",data[each])
    print("output:",s)
    print("accuracy:", np.mean(s==data[each]))
```
